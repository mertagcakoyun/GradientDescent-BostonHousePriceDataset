
Boston house prices dataset
---------------------------

**Data Set Characteristics:**  

    :Number of Instances: 506 

    :Number of Attributes: 13 numeric/categorical predictive. Median Value (attribute 14) is usually the target.

    :Attribute Information (in order):
        - CRIM     per capita crime rate by town
        - ZN       proportion of residential land zoned for lots over 25,000 sq.ft.
        - INDUS    proportion of non-retail business acres per town
        - CHAS     Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
        - NOX      nitric oxides concentration (parts per 10 million)
        - RM       average number of rooms per dwelling
        - AGE      proportion of owner-occupied units built prior to 1940
        - DIS      weighted distances to five Boston employment centres
        - RAD      index of accessibility to radial highways
        - TAX      full-value property-tax rate per 10,000 USD
        - PTRATIO  pupil-teacher ratio by town
        - B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
        - LSTAT    % lower status of the population
        - MEDV     Median value of owner-occupied homes in 1000 USD

    :Missing Attribute Values: None

    :Creator: Harrison, D. and Rubinfeld, D.L.

This is a copy of UCI ML housing dataset.
https://archive.ics.uci.edu/ml/machine-learning-databases/housing/

```python
#importing libraries and reading dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
     
bostonData = load_boston()
```

```python
data = pd.DataFrame(bostonData.data, columns=bostonData["feature_names"])
print(data)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =  train_test_split(data,  bostonData.target , test_size = 0.33, random_state = 0)
```

```python
#Determining the rate of test and train datas. Secondly fitting linear refression function.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(data, bostonData.target, test_size=0.33, random_state = 0)

linreg = LinearRegression().fit(X_train,y_train)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python

Y_pred = linreg.predict(X_test)
plt.figure(figsize=(20,10))
plt.scatter(y_test, Y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
```

```python
#getting MSE funtion from sklearn library
meanSquareError = sklearn.metrics.mean_squared_error(Y_test, Y_pred)
print(meanSquareError)
```

```python
#implementing gradient descent function
def GradientDescent(test,pred,learningRate,iteration=100,theta=100):
    thetaHistory = [x for x in range(iteration)]
    thetaHistory[0]=theta;
    for i in range(100):
        theta = theta - learningRate * sklearn.metrics.mean_squared_error(test, pred)
        thetaHistory[i]=theta;
        print(i,theta)
    plt.figure(figsize=(20,10))
    plt.scatter(thetaHistory, [x for x in range(iteration)])
```

```python
# running gradient descent fntion with different learning rates
GradientDescent(Y_test,Y_pred,0.5)
```

```python
GradientDescent(Y_test,Y_pred,0.1)
```

```python
GradientDescent(Y_test,Y_pred,0.02)
```
